```python
import pandas as pd
import numpy as np
```

```python
# Read the datafiles 
BMI_males = pd.read_csv("BMI_Males.csv")
BMI_women = pd.read_csv("BMI_Women.csv")
GDP = pd.read_csv("GDP.csv")
Diphtheria = pd.read_csv("Diphtheria.csv")
Hepatitis = pd.read_csv("Hepatitis.csv")
Measles = pd.read_csv("Measles.csv")
Polio = pd.read_csv("Polio.csv")
Life_expectancy = pd.read_csv("Life_Expectancy.csv")
population = pd.read_csv("population.csv")
Mortality_rate = pd.read_csv("Mortality_Rate.csv")
child_mortality = pd.read_csv("child_mortality.csv")
doctors = pd.read_csv("doctors.csv")
```

```python
# Clean the datasets 
BMI_males.head()

# Rename columns
BMI_males = BMI_males.rename(columns={"Entity": "Country"})

# Drop columns 
BMI_males = BMI_males.drop(columns=['Code'])

# Check for nulls columns
BMI_males.isna().sum()
BMI_males

```

```python
# Clean the datasets 
BMI_women.head()

# Rename columns
BMI_women = BMI_women.rename(columns={"Entity": "Country"})

# Drop columns 
BMI_women = BMI_women.drop(columns=['Code'])

# Check for nulls columns
BMI_women.isna().sum()
BMI_women
```

```python
# Clean the datasets 
GDP.head()

# Rename 
GDP = GDP.rename(columns={"Entity": "Country", "GDP per capita, PPP (constant 2017 international $)": "GDP Per Capita"})

# Drop columns 
GDP = GDP.drop(columns=['Code'])

# Check for nulls columns
GDP.isna().sum()
GDP


```

```python
# Clean the datasets 
Diphtheria.head()

# Isolate columns needed
Diphtheria = Diphtheria[["Period", "Location", "Value"]]
Diphtheria

# Rename columns
Diphtheria = Diphtheria.rename(columns={"Period": "Year", "Location": "Country", "Value":"Diphtheria"})

# Check missing values
Diphtheria.isna().sum()
Diphtheria

```

```python
# Clean the datasets 
Hepatitis.head()

# Isolate columns needed
Hepatitis = Hepatitis[["Period", "Location", "Value"]]
Hepatitis

# Rename columns
Hepatitis = Hepatitis.rename(columns={"Period": "Year", "Location": "Country", "Value":"Hepatitis"})

# Check missing values
Hepatitis.isna().sum()
Hepatitis
```

```python
# Clean the datasets 
Measles.head()

# Isolate columns needed
Measles = Measles[["Period", "Location", "Value"]]
Measles

# Rename columns
Measles = Measles.rename(columns={"Period": "Year", "Location": "Country", "Value":"Measles"})

# Check missing values
Measles.isna().sum()
Measles
```

```python
# Clean the datasets 
Polio.head()

# Isolate columns needed
Polio = Polio[["Period", "Location", "Value"]]
Polio

# Rename columns
Polio = Polio.rename(columns={"Period": "Year", "Location": "Country", "Value":"Polio"})

# Check missing values
Polio.isna().sum()
Polio
```

```python
Life_expectancy.head()

# Clean the datasets 
Life_expectancy.head()

# Isolate columns needed
Life_expectancy = Life_expectancy[["Period", "Location", "Life Expectancy"]]
Life_expectancy

# Rename columns
Life_expectancy = Life_expectancy.rename(columns={"Period": "Year", "Location": "Country"})

# Check missing values
Life_expectancy.isna().sum()
Life_expectancy
```

```python
# Clean the datasets 
population.head()

# Rename columns
population = population.rename(columns={"Entity": "Country", "Population (historical estimates)":"Population"})

# Drop columns 
population = population.drop(columns=['Code'])

# Isolate Years over 2000 
population = population[population['Year'] >= 2000]

# Check for nulls columns
population.isna().sum()
population
```

```python
# Clean the datasets 
Mortality_rate.head()

# Isolate columns needed
Mortality_rate = Mortality_rate[["Period", "Location", "FactValueNumeric"]]
Mortality_rate

# Rename columns
Mortality_rate = Mortality_rate.rename(columns={"Period": "Year", "Location": "Country", "FactValueNumeric":"Adult Mortality Rate"})

# Take the average of the 3 death rates for years
Mortality_rate = Mortality_rate.groupby(["Year", "Country"])["Adult Mortality Rate"].mean().reset_index()

# Check missing values
Mortality_rate.isna().sum()
Mortality_rate
```

```python
# Clean the datasets 
child_mortality.head()

# Clean the datasets 
child_mortality.head()

# Isolate columns needed
child_mortality = child_mortality[["Period", "Location", "FactValueNumeric"]]
child_mortality

# Rename columns
child_mortality = child_mortality.rename(columns={"Period": "Year", "Location": "Country", "FactValueNumeric":"Child Mortality Rate"})

# Check missing values
child_mortality.isna().sum()
child_mortality
```

```python
# Clean the datasets 
doctors.head()

# Clean the datasets 
doctors.head()

# Isolate columns needed
doctors = doctors[["Period", "Location", "Value"]]
doctors

# Rename columns
doctors = doctors.rename(columns={"Period": "Year", "Location": "Country", "Value":"Number of Doctors"})

# Check missing values
doctors.isna().sum()
doctors
```

```python
from functools import reduce

# List of your 10 datasets
datasets = [BMI_males, BMI_women, GDP, Diphtheria, Hepatitis, Measles, Polio, population, Mortality_rate, child_mortality, doctors]

# Merging all datasets based on "Year" and "Country"
merged_data = reduce(lambda left, right: pd.merge(left, right, on=["Year", "Country"], how="inner"), datasets)

# Display the resulting DataFrame
merged_data = merged_data.dropna()
merged_data
```

```python
# Convert to a CSV
merged_data.to_csv('Life-Expectancy-Data.csv', index=False)

```
